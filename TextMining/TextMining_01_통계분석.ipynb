```python
!pip install statsmodels
```

```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

```python
wine.head(3)
```

```python
wine.loc[wine['type'] == 'red', 'quality']
```

```python
red_wine_quality = wine.loc[wine['type'] == 'red', 'quality']
white_wine_quality = wine.loc[wine['type'] == 'white', 'quality']
```

```python
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
#  pvalue=8.168348870049682e-24 (<0.05)
# >> 두 집단(레드와인, 화이트 와인) 간 평균의 차이가 있음
# >> 귀무가설로 설정한 H0 : 두 집단간 평균의 차이는 없다 는 기각되고,
# >> 대립가설(연구가설)로 설정한 H1(Ha): 두 집단간 평균의 차이가 있다가 채택된다.
# (단, 일반화는 아님 : 전제조건)
```

```python
# 변수 사이의 상관관계를 히트맵(heatmap) 시각화하기

def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
titanic['age'][:3]
```

```python
titanic['age'].apply(category_age)
```

```python
titanic['age2'] = titanic['age'].apply(category_age)
```

```python
titanic.head(3)
```

```python
titanic.columns
```

```python
heatmap_data = titanic[['survived','sex','age2', 'pclass','fare']]
```

```python
heatmap_data[:3]
```

```python
heatmap_data.astype('float').corr()
```

```python
colormap = plt.cm.RdBu
# cm : colormap
sns.heatmap(heatmap_data.astype('float').corr(), linewidth=0.1, vmax=1.0,
            square=True, cmap=colormap, linecolor='white', annot=True,
            annot_kws= {'size': 10})

# vmax=1.0 color map 최대값 (상관계수의 최대값 1.0 지정, v: value)
# square (정사각형)
# annot_kws= {'size': 10}) 셀에 표시되는 숫자의 크기 설정 (크기: 10, kws: keyword size)

plt.show()
```
