Zero-shot Classification

```python
from transformers import pipeline

classfier = pipeline('zero-shot-classification')

classfier(
    'This is a course about the generative AI library',
    candidate_labels=['education', 'politics', 'business']
)

# 교육(education) 확률이 85.3% >> 이 문장은 교육관련이라고 판단
```

```python
from transformers import pipeline
import gradio as gr

classfier = pipeline('zero-shot-classification')

def classfier_text(text, labels):
  labels = labels.split(',')
  result = classfier(text, candidate_labels=labels)

  # 결과 딕셔너리 생성
  result_dict = {}
  for i in range(len(result['labels'])):
    result_dict[result['labels'][i]] = result['scores'][i]

  return result_dict

iface = gr.Interface(
    fn=classfier_text,
    inputs=[
        gr.Textbox(label='Text to classify'),
        gr.Textbox(label='Candidate Labels')
    ],
    outputs=gr.Label(label='classification Result', num_top_classes=3)
)

iface.launch(share=True)
```

```python
from transformers import pipeline
import gradio as gr
# NER pipeline 초기화
ner = pipeline("ner", grouped_entities=True)

def extract_entities(text):
  result = ner(text)

  # 코드 수정
  entities = {}
  for entity_group in result:
    entities[entity_group['word']] =  [entity_group['entity_group'],entity_group['score']]
  print(entities)

  return entities

iface = gr.Interface(
    fn=extract_entities,
    inputs=gr.Textbox(label='Text for NER'),
    outputs=gr.Textbox(label='result')
)
iface.launch(share=True)
```

질의응답 (QA)

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do you work?",
    context="My name is Jin-wook and I work at Hansikyung in Seoul.",
)
```

```python
from transformers import pipeline
import gradio as gr

question_answerer = pipeline("question-answering")

def answer_question(question, context):
  result = question_answerer(question=question, context=context)
  return (result['answer'], result['score'])

iface = gr.Interface(
    fn=answer_question,
    inputs=[
        gr.Textbox(label='Question'),
        gr.Textbox(label='Context')
    ],
    outputs=[
        gr.Textbox(label='Answer'),
        gr.Textbox(label='Confidence')
    ]
)
iface.launch(share=True)
```
