```python
# BMI 계산기 1 - input 여러 개 다루기

import gradio as gr

def bmi(name, height, weight):
  bmi_val = round(weight / ((height /100) ** 2),2)
  return (f'안녕하세요. {name}님의 BMI는 {bmi_val}입니다.')

interface = gr.Interface(fn=bmi,
             inputs=['text', 'number', 'number'],
             outputs='text')

interface.launch()
```

```python
# BMI 계산시 2 - output 여러 개 다루기

import gradio as gr

def bmi(name, height, weight):
  bmi_val = round(weight / ((height /100) ** 2),2)
  emotion = ''
  if bmi_val >= 23:
    emotion = '😢'
  else:
    emotion = '😊'

  return (f'안녕하세요. {name}님의 BMI는 {bmi_val}입니다.', emotion)

interface = gr.Interface(fn=bmi,
             inputs=['text', 'number', 'number'],
             outputs=['text','text'])

interface.launch()
```

```python
# BMI 계산시 3 - components 사용하기

import gradio as gr

def bmi(name, height, weight):
  bmi_val = round(weight / ((height /100) ** 2),2)
  emotion = ''
  if bmi_val >= 23:
    emotion = '😢'
  else:
    emotion = '😊'

  return (f'안녕하세요. {name}님의 BMI는 {bmi_val}입니다.', emotion)

interface = gr.Interface(fn=bmi,
             inputs=['text', gr.Slider(0, 300, label='키(cm)'), gr.Slider(0, 150, label='몸무게(kg)')],
             outputs=['text','text'])

interface.launch()
```

```python
# BMI 계산시 4 - checkbox components 사용하기

import gradio as gr

def bmi(name, height, weight, diet):
  bmi_val = round(weight / ((height /100) ** 2),2)
  emotion = ''
  if bmi_val >= 23:
    emotion = '😢'
  else:
    emotion = '😊'

  if diet:
    diet = '다이어트 중'
  else:
    diet = '살찌는 중...'

  return (f'안녕하세요. {name}님의 BMI는 {bmi_val}입니다.', emotion, diet)

interface = gr.Interface(fn=bmi,
             inputs=['text',
                     gr.Slider(0, 300, label='키(cm)'),
                     gr.Slider(0, 150, label='몸무게(kg)'),
                     gr.Checkbox(label='다이어트 여부')],
             outputs=['text','text', 'text'])

interface.launch()
```

```python
import gradio as gr

def bmi(name, height, weight, diet):
    bmi_val = round(weight / ((height / 100) ** 2), 2)
    emotion = ''
    if bmi_val >= 23:
        emotion = '😢'
    else:
        emotion = '😊'

    if diet:
        diet = '다이어트 중'
    else:
        diet = '살찌는 중...'

    return (f'안녕하세요. {name}님의 BMI는 {bmi_val}입니다.', emotion, diet)

interface = gr.Interface(fn=bmi,
                         inputs=['text',
                                 gr.Slider(0, 300, label='키(cm)'),
                                 gr.Slider(0, 150, label='몸무게(kg)'),
                                 gr.Checkbox(label='다이어트 여부')],
                         outputs=['text', 'text', 'text'],
                         flagging_options=['good', 'bad', 'error'],
                         description='문제가 생기면 Flag를 누르세요.',
                         examples=[['김건우', 180, 90, True], ['사나', 167, 47, False]]
)

interface.launch()
```
