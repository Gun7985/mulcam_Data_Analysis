```python

openai.api_key = OPENAI_API_KEY

model = 'gpt-3.5-turbo'

query = '''
나의 MBTI : intj
상대방 이름 : 카리나
관계 : 연인
상황 : 만난 지 100일 된 상황
질문 : 100일 기념 선물로 무엇이 좋을 것 같아? 보통 어떤 거를 선물하지?
'''
# few shot

message = [
    {'role':'system', 'content':'연애전문상담가'},
    {'role':'user', 'content':query}
]

try:
  response = openai.ChatCompletion.create(
      model = model,
      messages = message,
      temperature = 0.2
      # 0 에 가까울수록 일반적인 답변
      # 0-1 1에 가까우면 창의적 답변(Hallucination(환각,거짓) 가능성 존재)
  )
  # print(response)
  print(response.choices[0].message['content'])

except Exception as e:
  print(f'에러가 발생했어요.:{e}')
```

```python
my_mbti = input('내 MBTI: ').upper()
your_mbti = input('상대방 MBTI: ').upper()
your_name =input('상대방 이름:' )
relation = input('관계: ')
situation = input('상황: ')
question = input('질문: ')
query = f'''
나의 MBTI : {my_mbti}
상대방 MBTI : {your_mbti}
상대방 이름 : {your_name}
관계 : {relation}
상황 : {situation}
질문 : {question}
'''
messages = [
    {'role': 'system', 'content': '연예전문상담가'},
    {'role': 'user', 'content': query},
]
print('답변을 준비중입니다.... 잠시만 기다려 주실거죵~^^')
response = openai.ChatCompletion.create(
        model = model,
        messages=messages,
        temperature=0.7
        # 0 에 가까울 수록 일반적인 답변.
        # 0-1 1에 가까우면 창의적 답변(할루시네이션 가능성 존재)
    )
answer = response.choices[0].message['content']
messages.append({'role': 'assistant', 'content':answer})
print(answer)

# 질문이 끊기지 않고 계속 하도록 while 사용 (multi-turn 구현)

while True:
  # 사용자로부터 질문을 받습니다.
  question = input('질문:')
  query = f'Q:[question]'
  # 질문을 메세지 배열에 추가합니다
  messages.append({'role':'user', 'content':query})
  # chatGPT API 호출 >> 응답 받기
  print('답변을 준비중입니다...')

  response = openai.ChatCompletion.create(
        model = model,
        messages=messages,
        temperature=0.7
        # 0 에 가까울 수록 일반적인 답변.
        # 0-1 1에 가까우면 창의적 답변(할루시네이션 가능성 존재)
    )
  answer = response.choices[0].message['content']
  # 응답을 출력합니다.
  print('=='*40)

  messages.append({'role': 'assistant', 'content':answer})
  print(answer)

  # 사용자에게 계속 질문할 건지 확인하기
  while True:
    q = input('더 질문하실건가요?(y/n): ').lower() # 입력된 영어문자를 전부 소문자로 변환
    if q in ['y','n','ㅛ','ㅜ']:
      break # 정지조건
    else:
      print('유효한 입력이 아닙니다. y나 n 중 선택하여 입력해주세요.')

  if q in ['n','ㅜ']:
    print('종료합니다.')
    break

```
