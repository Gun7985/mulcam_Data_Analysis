```python
# 정규화 (Scailing)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_tr)

x_tr = scaler.transform(x_tr)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)
```

```python
# 모델링 KNN

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(x_tr, y_tr.values.ravel())
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=10)
dt.fit(x_tr, y_tr)
```

```python
# 모델 평가
pred_val = knn.predict(x_val)
```

```python
pred_val
```

```python
pred_val_proba_knn = knn.predict_proba(x_val)
pred_val_proba_dt = dt.predict_proba(x_val)
```

```python
pred_val_proba_knn[:3]
```

```python
# roc_auc

from sklearn.metrics import roc_auc_score

score_knn = roc_auc_score(y_val, pred_val_proba_knn[:, 1])
score_dt = roc_auc_score(y_val, pred_val_proba_dt[:, 1])

print(score_knn, score_dt)
```

```python
best_model = None
best_score = 0

for i in range(2, 10):
    model = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
    model.fit(x_tr, y_tr.values.ravel())
    pred_val_proba = model.predict_proba(x_val)
    score = roc_auc_score(y_val,pred_val_proba [:, 1])
    print(i, '개의 이웃 확인: ', score)
    if best_score <= score:
        best_model = model
        best_score = score
```
