```python
from sklearn.preprocessing import LabelEncoder

x = pd.concat([x_train, x_test], axis=0)

for col in col_cat:
    le = LabelEncoder()
    le.fit(x[col])
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
```

```python
# 데이터 정규화-데이터 스케일링

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_tr[col_num] = scaler.fit_transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
x_test[col_num] = scaler.transform(x_test[col_num])
```

```python
y_tr
```

```python
# 선형모델 사용하지 않음 (LogisticRegressor 사용 안할 결심)
# 라벨인코딩을 했기 때문에 >> 강제로 범주형 데이터로 만들어서

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_tr, y_tr.values.ravel())
```

```python
'''
-----------------------실무필수항목------------------------------
1. n_estimators : 결정트리 (default=100)
2. max_depth : 트리 깊이 최대값 (default=None)
3. max_features : 최적의 분할을 위해 고려할 최대 feature 개수 (default=None)
-------------------------세부항목---------------------------------
4. min_samples_split : node 분리에 필요한 최소 샘플 수 (default=2)
5. min_samples_leaf : leaf 노드에 필요한 최소 샘플 수 (default=1)
'''
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(x_tr, y_tr.values.ravel())
```

```python
'''
-----------------------실무필수항목------------------------------
1. n_estimators : boosting할 결정트리 개수(default=100)
2. max_depth : 트리 깊이 최대값 (default=None)
3. learning_rate : 학습률 (default=0.1)
4. lambda : L2 규제 (default=1)
5. alpha : L1 규제 (default=0)
-------------------------세부항목---------------------------------
6. min_child_weight : 자식노드에 필요한 가중치 합의 최솟값 (default=1)
7. subsample : 각 트리에 사용할 데이터 샘플의 비율 (0.5-~) (default=1)
8. colsample_bytree : 각 트리 생성시, 사용되는 feature(column 열) 비율 의미 (default=1)

'''
```

```python
xgb_tune = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=10)
xgb_tune.fit(x_tr, y_tr.values.ravel(), early_stopping_rounds=50,
        eval_metric='auc', eval_set=[(x_val, y_val)], verbose=10)

# early_stopping_rounds=50 조기종료
# eval_metric='auc' 평가(cvaluation) 지표 ayc (1에 가까울수록 )
```

```python
# roc_auc

from sklearn.metrics import roc_auc_score

rf.predict(x_val)
```

```python
pred_rf = rf.predict_proba(x_val)
pred_xgd= xgb.predict_proba(x_val)
pred_xgb_tune = xgb_tune.predict_proba(x_val)
```

```python
score_rf = roc_auc_score(y_val, pred_rf[:,1])
score_xgd = roc_auc_score(y_val, pred_xgd[:,1])
score_xgb_tune = roc_auc_score(y_val, pred_xgb_tune[:,1])
```
