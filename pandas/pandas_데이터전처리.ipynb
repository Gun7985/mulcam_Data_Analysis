```python
# 데이터 변환

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 한국인, 일본인 각 성인 100명 육류 소비량 데이터 생성

np.random.randn(1000) # 0~1 사이 사상의 데이터 1000개 생성

meat_consumption_kor = 5 * np.random.randn(1000) + 54
meat_consumption_jpn = 4 * np.random.randn(1000) + 32

# 5 : 가상추세 / 54 : 가상평균값
```

```python
# 데이터 프레임 생성
pd.DataFrame({'한국인':meat_consumption_kor,
              '일본인':meat_consumption_jpn})
```

```python
raw = pd.DataFrame({'한국인':meat_consumption_kor,
              '일본인':meat_consumption_jpn})


df = raw.copy()
df.head(3)
```

```python
# 한국인 육류 소비량 히스토그램

plt.hist(meat_consumption_kor)
plt.xlabel('Korea')
plt.show()
```

```python
# 일본인 육류 소비량 히스토그램

plt.hist(meat_consumption_jpn)
plt.xlabel('Japan')
plt.show()
```

```python
# 표준 표준화 함수 zscoer() 함수
# 기존 데이터프레임 데이터에 새 열 생성(zscore 즉 데이터표준화 한 컬럼 추가)

import scipy.stats as ss

df['한국인 표준화'] = ss.zscore(meat_consumption_kor)
df['일본인 표준화'] = ss.zscore(meat_consumption_jpn)
```

```python
df.head(3)
```

```python
# 표준표준화 수식 활용
# (x-x_bar) / sigma >> zscore

df['한국인 표준화_수식'] =\
(meat_consumption_kor - np.mean(meat_consumption_kor)) / np.std(meat_consumption_kor)

df['일본인 표준화_수식'] =\
(meat_consumption_jpn - np.mean(meat_consumption_jpn)) / np.std(meat_consumption_jpn)
```

```python
df.head(3)
```

```python
# 표준화된 데이터에 대한 히스토그램 그리기

plt.hist(df['한국인 표준화'])
plt.xlabel('Korea')
plt.show()

plt.hist(df['일본인 표준화'])
plt.xlabel('Japan')
plt.show()
```

```python
# 사이킷런 스케일러 활용 표준화 (*** 암기할 것)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['한국인 표준화_사이킷런_표준화'] = scaler.fit_transform(df[['한국인']])
df['일본인 표준화_사이킷런_표준화'] = scaler.fit_transform(df[['일본인']])
```

```python
df.head(3)
```

```python
# 정규화 : min-max (min : 0, max : 1)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['한국인 정규화_사이킷런_최소최대'] = scaler.fit_transform(df[['한국인']])
df['일본인 정규화_사이킷런_최소최대'] = scaler.fit_transform(df[['일본인']])
```

```python
df.head(3)
```

```python
# min-max 변환 수식 이용
# (x - min) / (max-min) = (x - 0) / (1-0)

min_ = np.min(meat_consumption_kor)
max_ = np.max(meat_consumption_kor)
df['한국인_수식_최대최소']= (df[['한국인']] - min_) / (max_ - min_)

min_ = np.min(meat_consumption_jpn)
max_ = np.max(meat_consumption_jpn)
df['일본인_수식_최대최소']= (df[['일본인']] - min_) / (max_ - min_)

df.head(3)
```

```python
# 정규분포
# 평균 50, 표준편차(std) : 10 인 데이터 200개 생성

data = 10 * np.random.randn(200) + 50
data[:10]
```

```python
# 분포 시각화
plt.hist(df['값'], bins=20, rwidth=0.9)
plt.show()
```

```python
plt.boxplot(df['값'])
plt.show()
```

```python
Q1 = df['값'].quantile(.25)
Q2 = df['값'].quantile(.50)
Q3 = df['값'].quantile(.75)

IQR = Q3 - Q1

print('Q1 = ',Q1)
print('Q2 = ',Q2)
print('Q3 = ',Q3)
print('IQR = ',IQR)
```

```python
df_desc = df.describe()
df_desc.index
```

```python
df_desc.loc['25%',:]
```

```python
Q1 = df['값'].describe()['25%']
Q2 = df['값'].describe()['50%']
Q3 = df['값'].describe()['75%']

IQR = Q3 - Q1

print('Q1 = ',Q1)
print('Q2 = ',Q2)
print('Q3 = ',Q3)
print('IQR = ',IQR)
```

```python
condition = df['값'] > (Q3 + IQR * 1.5)
df[condition]
upper_outlier = df[condition]
print(upper_outlier)
```

```python
condition = df['값'] < (Q1 - IQR * 1.5)
df[condition]
lower_outlier = df[condition]
print(lower_outlier)
```
