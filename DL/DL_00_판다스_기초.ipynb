```python
# 새로운 열 만들기

df =\
pd.DataFrame(
    {"a": [4,5,6,7],
     "b": [8,9,10,11],
     "c": [12,13,14,15]},
    index=pd.MultiIndex.from_tuples(
        [('d',1), ('d', 2), ('e',1),('e',2)]
    )

)

df
```

```python
# 조건에 맞는 새 열 만들기

df['sum'] = df['a'] + df['b'] + df['c']
df
```

```python
# assign() 이용, 조건에 맞는 새열 만들기

# a열, b열, c열 값을 모두 곱해서 d열 생성

df = df.assign(multiply= lambda df: df['a']*df['b']*df['c'])
df
```

```python
# 숫자형 데이터를 구간 나누기

# a 열을 두개로 나누어 각각 새롭게 레이블 만드는 법
df['qcut'] = pd.qcut(df['a'], 2, labels=['600이하', '600이상'])
df
```

```python
# 기준 값 이하와 이상을 모두 통일 시키기

# a열에서 5 이하는 모두 5로, 6이상은 모두 6으로 변환

df['clip'] = df['a'].clip(lower=5, upper=6)
```

```python
# 행과 열 변환

df=\
pd.DataFrame({'A': {0:'a', 1:'b', 2:'c'},
              'B': {0:1, 1:3, 2:5},
              'C': {0:2, 1:4, 2:6},


              })

df
```

```python
# 데이터 병합 (merge)

df1 = pd.DataFrame( {
    "x1": ['a','b','c'],
    'x2':[1,2,3]
})

df1
```

```python
df2 = pd.DataFrame( {
    "x1": ['a','b','d'],
    'x3':['T','F','T']
})

df2
```

```python
df3 = pd.DataFrame( {
    "x1": ['b','c','d'],
    'x2':[2,3,4]
})

df3
```

```python
# 왼쪽 열을 축으로 병합(merge)

pd.merge(df1, df2, how='left', on='x1')
```

```python
# 오른쪽 열을 축으로 병합(merge)

pd.merge(df1, df2, how='right', on='x1')
```

```python
# 공통 값만 병합(merge)

pd.merge(df1, df2, how='inner', on='x1')
```

```python
# 모든 값만 병합(merge)

pd.merge(df1, df2, how='outer', on='x1')
```

```python
# 특정 열 비교, 공통값이 존재하는 경우만 가져오기

df1[df1.x1.isin(df2.x1)]
```

```python
# 특정 열 비교, 공통값이 존재하는 경우, 해당 값 제외하고 병합

df1[~df1.x1.isin(df2.x1)]
```

```python
# 공통 값이 있는 것만 병합
pd.merge(df1, df3) # inner join(default)
```

```python
# 모두 병합
pd.merge(df1, df3, how='outer')
```

```python
# 어디서 병합되었는지 원천(source) 표시하기

pd.merge(df1, df3, how='outer', indicator=True)
```

```python
# 원하는 병합만 남기기
pd.merge(df1, df3, how='outer', indicator=True).query('_merge=="left_only"')
```

```python
# merge 컬럼 없애기
pd.merge(df1, df3, how='outer', indicator=True).query('_merge=="left_only"').drop(columns=['_merge'])
```
